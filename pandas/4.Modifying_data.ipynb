```python
import pandas as pd
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
# creating a column in a file or data and assigning a value

df['Saim']=100
df
# column will be locate at the end
```

```python
# to set its loc 
df.insert(loc = 0, column = 'Marks' , value = 100)
df
```

## Delete

```python
# for drop a specific column needs to tell the axis    (x for row  , 1 for column)
df.drop("Saim",axis=1)
```

```python
df.head()
```

**drop does not delte permanently anything it just makes another copy if we want to delete a column permanent then we have to use inplace = True**

```python
df.drop("Saim",axis=1 , inplace=True)
```

```python
df.head()
```

```python
# deleting through row
df.drop(2)
```

## Rename

```python
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4], 'C': [5, 6]})
df = df.set_axis(['Alpha', 'Beta', 'Gamma'], axis=1)        #setting names
df
```

```python
df.rename(columns = {'Alpha' : 'A' , 'Beta':'B' , 'Gamma':'G'})
```

<br>

## Replace (specific values)

```python
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4], 'C': [5, 6]})
df
```

```python
df.replace(to_replace=1,
           value=5)
```

```python
# multiple value replace
df.replace(to_replace=[1,3,5],
           value=[10,10,10])
```
