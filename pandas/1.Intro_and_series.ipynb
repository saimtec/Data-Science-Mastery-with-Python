>## Series
**A Series is a one-dimensional labeled array in Pandas.**

```python
import pandas as pd

# creating series
s = pd.Series([1,2,3,4])
s
```

```python
# custom indexing
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print(s)
```

```python
# form a dictionary
data = {
    "Ali": 50, 
    "Sara": 80,
    "Ahmed": 65
    }
s = pd.Series(data)
print(s)
```

```python
s['Ali']       #accessing specific value
```

```python
s.size        # number of elements
```

```python
s.index       # index labels
```

```python
s.values      # underlying array
```

```python
s.dtype       # data type
```

```python
s.sum()       # sum
```

```python
s.mean()      # average
```

```python
s.max()       # max
```

```python
s.min()       # min
```

```python
s[s > 50]    #filtering
```

```python
s[1:2]    # Slicing
```
