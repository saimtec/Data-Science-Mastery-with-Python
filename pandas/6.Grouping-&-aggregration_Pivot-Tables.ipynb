### Grouping and aggregation are used because they help you summarize, understand, and analyze the data
* How many?

* What is the average?

* What is the total?

* What is the maximum/minimum?

* How do different groups compare?

**Grouping separates your dataset into categories.**
**Aggregation summarizes each category.**

```python
import pandas as pd
df=pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
# Average age of male vs female
df.groupby('Sex')['Age'].mean()
```

```python
# Survival rate by gender
df.groupby('Sex')['Survived'].mean()
```

### Group by 2 columns

```python
# Survival rate of gender within each class:
df.groupby(['Pclass','Sex'])['Survived'].mean()
```

```python
# Aerage age in each class according to gender
df.groupby(['Pclass','Sex'])['Age'].mean()
```

### Named aggregations

```python
df.groupby('Sex').agg(
    avg_age=('Age','mean'),
    survival_rate=('Survived','mean'),
    avg_fare=('Fare','mean')
)
```

```python
df.groupby(['Pclass','Sex']).agg(
    avg_age=('Age','mean'),
    survival_rate=('Survived','mean'),
    avg_fare=('Fare','mean')
)
```

## Pivot Tables

```python
# (Gender on rows, Class on columns)
df.pivot_table(
    values='Survived',
    index='Pclass',
    columns='Sex',
    aggfunc='mean'
)
```

```python
# Average fare by class and embark point
df.pivot_table(
    values='Fare',
    index='Pclass',
    columns='Embarked',
    aggfunc='mean'
)
```

```python
# Count how many passengers per class per gender
df.pivot_table(
    values='PassengerId',
    index='Pclass',
    columns='Sex',
    aggfunc='count'
)
```

```python
# Fill missing values
df.pivot_table(
    values='Fare',
    index='Pclass',
    columns='Embarked',
    aggfunc='mean',
    fill_value=0
)
```
