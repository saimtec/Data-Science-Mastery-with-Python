># Covariance

* Measures relationship between two variables.
* tells the direction not the strength
* if var2 increases with var2 it is +covariance , if decreses it is -covariance

**Formula (Covariance):**
$$\text{Cov}(X,Y) = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})
$$

**Where:**

$\text{Cov}(X,Y)$ = Sample covariance

$n$ = Sample size (number of observations)

$X_i, Y_i$ = The $i$-th observations

$\bar{X}, \bar{Y}$ = Sample means

># Corelation
* Describe the strength between two variables

**Formula (Corelation):**
$$r = \frac{\text{Cov}(X,Y)}{\sigma_X \sigma_Y}
$$

**Where:**

$r$ = Sample Pearson Correlation Coefficient

$\text{Cov}(X,Y)$ = Sample covariance of $X$ and $Y$

$\sigma_X, \sigma_Y$ = Sample standard deviations of $X$ and $Y$

$r \in [-1, 1]$, 

where:
$r = 1$: Perfect positive linear relationship

$r = -1$: Perfect negative linear relationship

$r = 0$: No linear relationship
