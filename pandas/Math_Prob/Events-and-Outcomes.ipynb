># Probability
* Tells us how likely somethings happen
$$
P(\text{Event}) = 
\frac{\text{Number of favorable outcomes}}
{\text{Total number of possible outcomes}}
$$


<br><br>

> # What Is an Outcome?

* An outcome is a single possible result of an experiment.

**Examples:**

* You toss a coin → possible outcomes = Heads, Tails

* You roll a dice → outcomes = 1, 2, 3, 4, 5, 6

* You draw a card → outcomes = 52 different cards

**Outcome = one result.**

<br><br>

## What Is a Sample Space (Ω)?

* Sample space = set of all possible outcomes.

**Examples:**

* Coin toss:
Ω = {H, T}

* Dice roll:
Ω = {1, 2, 3, 4, 5, 6}

* Two coin tosses:
Ω = {HH, HT, TH, TT}

<br><br>

># What Is an Event?

* An event is a group/set of outcomes we are interested in.

**Example (rolling dice):**

* Event “even number” = {2, 4, 6}

* Event “number > 4” = {5, 6}

* Event “prime number” = {2, 3, 5}


<br>

## Types of Events

**1 Simple Event**

Only one outcome.

{H}

{4}

**2️ Compound Event**

More than one outcome.

{2, 4, 6}

{H, T}

**3️ Impossible Event**

Cannot happen.

Rolling a 7 on one dice → Probability = 0


**4️ Certain Event**

Must happen.

Getting a number 1–6 on a dice → Probability = 1


**5️ Mutually Exclusive Events**

Both cannot happen together.
Example:

Rolling 2 and 5 together → Impossible

Getting Heads and Tails on a single toss → Impossible


**6️ Complementary Events**

Opposites.
Example:

Heads & Tails
If you don’t get Heads → you must get Tails.


<br><br>

*Event A = “get even number” = {2, 4, 6}

P(\text{A}) = 
\frac{\text{3}}

**Example 2: Coin toss**
