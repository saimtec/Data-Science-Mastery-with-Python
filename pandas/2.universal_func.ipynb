## Merge
* pd.merge() function is used to combine two different DataFrames
* How to merge (inner, left, right, outer)

```python
data_users = {
    'UserID': [101, 102, 103, 104],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'City': ['NY', 'LA', 'NY', 'SF']
}
users = pd.DataFrame(data_users)
users
```

```python
data_orders = {
    'User_ID_FK': [102, 101, 104, 105], 
    'Product': ['Mouse', 'Laptop', 'Keyboard', 'Monitor'],
    'Price': [25, 1200, 75, 300]
}
orders = pd.DataFrame(data_orders)
orders
```

```python
merged = pd.merge(
    users,            
    orders,            
    left_on='UserID',     
    right_on='User_ID_FK', 
    how='inner'            
)

print(merged)
```

## Join
* .join() method primarily combines DataFrames based on their indexes

```python
users.join(orders , how ='left')
```
