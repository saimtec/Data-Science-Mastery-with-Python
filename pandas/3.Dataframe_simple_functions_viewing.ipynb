**Loc and Iloc**

* loc is label based selection first parameter is for rows and second is for column
* Iloc is integer or number bases selection same  first parameter is for row and second is fpr column

```python

df.loc[1,'Age']     
```

```python
df.loc[1:7,['Age','Name']]
```

```python
df.iloc[1,5]
```

```python
df.iloc[1:8,[5,3]]
```
