## Data cleaning has some major topics:

 * Missing Values

* Duplicates

* Fixing Wrong Data Types

* Fixing Wrong / Incorrect Values

* String Cleaning & Formatting

```python
import pandas as pd
df=pd.read_csv("Titanic-Dataset.csv")
df
```

>### Check missing values
* NAN -> not a number 
* Null -> any empty thing can be a string 
>**so null is a broader view which contain both string as well as numbers but nan related only to numbers**

```python
df.isnull()   #checking for null
```

```python
df.isnull().sum()
```

### Remove rows with missing values

```python
df.dropna()
```

```python
#Remove rows where specific column is NaN
df.dropna(subset=['PassengerId'])
```

```python
# Remove columns that have NaN
df.dropna(axis=1)
```

<br><br>

>## Fill missing values

```python
# df.fillna(0)
df.fillna('Saim')
```

```python
# filling with mean , median etc
df['Age']=df['Age'].fillna(df['Age'].mean())
df
```

<br><br><br>

>## DUPLICATES

### Find duplicates

```python
df.duplicated()
df.duplicated().sum()
```

```python
# dropping duplicates
df.drop_duplicates()       #use inplace for permanent drop
```

### Remove duplicates based on specific columns

```python
df.drop_duplicates(subset=['Name'], inplace=True)
```

<br><br><br>

>## FIX WRONG DATA TYPES

```python
# check type
df.dtypes
```

```python
# convert type
df['Age'] = df['Age'].astype(int)
df['Fare'] = df['Fare'].astype(int)
```

```python
df.dtypes
```

```python
df['Age'] = df['Age'].astype(str)
df['Fare'] = df['Fare'].astype(float)
```

```python
df.dtypes
```

```python
# convert to string
df['Name'] = df['Name'].astype(str)
df.dtypes
```

<br><br><br>

>## STRING CLEANING & FORMATTING

```python
# Remove spaces
df['Name'] = df['Name'].str.strip()
```

```python
# Convert to lowercase / uppercase
df['Name'] = df['Name'].str.upper()
```

```python
df.head()
```

```python
df['Sex']=df['Sex'].str.replace('[^0-9]','',regex=True)
df
```
