# Distributions

**Distributions describe how data points are spread across values.**
<br><br>

>## Normal Distribution (Gaussian)
* A continuous, bell-shaped, symmetric distribution.Mean = Median = Mode.

![Bell Curve](Normal_Distribution.jpeg)


##  Probability Density Function (PDF)

The formula for the Normal Distribution's PDF is:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} 
\; e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

**Where:**

- x  = data value  
- $\mu  $ = mean  
- $\sigma $ = standard deviation  


---

## Empirical Rule (68–95–99.7 Rule)

- **68%** of data → \( $\mu \pm 1\sigma $\)  
- **95%** of data → \( $\mu \pm 2\sigma $\)  
- **99.7%** of data → \( $\mu \pm 3\sigma $\)  


---

#  Standard Normal Distribution

A special case of normal distribution where:
- Mean:  $\mu=0, \ $
- Standard deviation: $\sigma=1$ 

---

##  Standard Normal PDF

$$
\phi(z) = \frac{1}{\sqrt{2\pi}} \; e^{-\frac{z^2}{2}}
$$

Where:

- \( z \) = standardized value  




#  Z-Score Formula

Converts any raw score \( X \) to standard normal form:

$$
Z = \frac{X - \mu}{\sigma}
$$

**Meaning:**

- \( Z = 0 \) → exactly at the mean  
- \( Z = +1 \) → one SD above mean  
- \( Z = -1 \) → one SD below mean  


---

## Solved Problem Example

The scores on an exam follow a normal distribution with $\mu=70, \; \sigma=10$. A student scored 85. Find the Z-score and interpret it.Step 1: Apply the Z-score formula$$Z = \frac{X - \mu}{\sigma} = \frac{85 - 70}{10}$$Step 2: Simplify$$Z = \frac{15}{10} = 1.5$$Interpretation: The student's score of 85 is 1.5 standard deviations above the mean.

># Uniform Distribution:
* Key ConceptsA probability distribution where every value in a specified interval $[a, b]$ is equally likely.

* The probability density is constant across the interval.

* The graph is a characteristic rectangle shape.

**Probability Density Function (PDF)**
For a continuous uniform distribution on the interval $[a,b]$:$$f(x) =
\begin{cases}
\frac{1}{b - a}, & a \le x \le b \\
0, & \text{otherwise}
\end{cases}$$

Parameters:
- $a$ = minimum value
- $b$ = maximum value
 
The average is the midpoint of the interval:$$\mu = \frac{a + b}{2}$$Variance$$\sigma^2 = \frac{(b - a)^2}{12}$$
## Cumulative Distribution Function (CDF)
The CDF, $F(x)$, gives the probability that a random variable $X$ is less than or equal to $x$:$$F(x) =
\begin{cases}
0, & x < a \\
\frac{x - a}{b - a}, & a \le x \le b \\
1, & x > b
\end{cases}$$
## Solved Problem Example
The variable $X$ is uniformly distributed between $a=10$ and $b=30$.
Step 1: Find the Probability Density Function (PDF) Value $f(x)$
The constant height of the distribution is calculated as:$$f(x) = \frac{1}{b - a}$$
Substituting the values:$$f(x) = \frac{1}{30 - 10} = \frac{1}{20} = 0.05$$
Step 2: Find the Mean $\mu$The mean is the midpoint of the interval:$$\mu = \frac{a + b}{2}$$
Substituting the values:$$\mu = \frac{10 + 30}{2} = \frac{40}{2} = 20$$
Step 3: Find Probability $P(15 < X < 22)$
The probability is the ratio of the sub-interval's length to the total length:$$P(x_1 < X < x_2) = \frac{x_2 - x_1}{b - a}$$
Substituting the values:$$P(15 < X < 22) = \frac{22 - 15}{30 - 10}$$ $$P(15 < X < 22) = \frac{7}{20} = 0.35$$

># Binomial Distribution:
The Binomial Distribution is a discrete probability distribution that models the number of successes ($x$) in a fixed number of independent trials ($n$), where each trial has the same probability of success ($p$).

## Conditions for a Binomial Distribution (B.I.N.S.)
- Binary: Each trial has only two outcomes (Success/Failure).
- Independent: The outcome of one trial does not affect the others.
- Number of Trials ($n$): The number of trials must be fixed in advance.
- Same Probability ($p$): The probability of success ($p$) is the same for every trial.

## Probability Mass Function (PMF)
- The PMF gives the probability of getting exactly $x$ successes in $n$ trials:
$$P(X = x) = \binom{n}{x} p^x (1 - p)^{n-x}$$

**Where:**

$\binom{n}{x}$ is the binomial coefficient, often written as $_nC_x$.
- $n$ = number of trials
- $x$ = number of successful outcomes (0, 1, 2, ..., $n$)
- $p$ = probability of success on a single trial
- $1-p$ (or $q$) = probability of failure on a single trial

## Binomial Coefficient Formula
The number of ways to choose $x$ successes from $n$ trials:
$$\binom{n}{x} = \frac{n!}{x!(n-x)!}$$

## Mean, Variance, and Standard Deviation
These parameters characterize the shape and spread of the distribution:
* Mean (Expected Value)
This is the average number of successes you expect:
$$\mu = E(X) = n p$$

* Variance
This measures the spread of the distribution:
$$\sigma^2 = n p (1 - p)$$

* Standard Deviation
$$\sigma = \sqrt{n p (1 - p)}$$

## Solved Problem Example

- Problem:
 A manufacturing process produces defective parts at a rate of 5% ($p=0.05$). If we randomly select 10 parts 
 ($n=10$), what is the probability that exactly 2 of them ($x=2$) are defective?
 
 Step 1: Identify Parameters
 
 - $n = 10$ (Number of trials/parts)
 - $x = 2$ (Number of successes/defective parts)
 - $p = 0.05$ (Probability of success/defective)
 - $1-p = 0.95$ (Probability of failure/non-defective)
 
 Step 2: Calculate the Binomial Coefficient
 
 - We need to find the number of ways to choose 2 defective parts out of 10:
 $$\binom{10}{2} = \frac{10!}{2!(10-2)!} = \frac{10 \times 9}{2 \times 1} = 45$$
 
 
 Step 3: Apply the PMF Formula
 
 Substitute the values into the probability formula $P(X = x) = \binom{n}{x} p^x (1 - p)^{n-x}$:
 
 $$P(X = 2) = \binom{10}{2} (0.05)^2 (0.95)^{10-2}$$
 
 $$P(X = 2) = 45 \times (0.0025) \times (0.95)^8$$
 
 $$P(X = 2) \approx 45 \times 0.0025 \times 0.6634$$
 
 $$P(X = 2) \approx \mathbf{0.0746} \quad (\text{or } 7.46\%)$$
