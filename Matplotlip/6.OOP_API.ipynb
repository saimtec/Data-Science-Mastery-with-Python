```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Name": ["Ali", "Sara", "Zain", "Ahmed", "Hina"],
    "Age": [22, 24, 21, 27, 23],
    "Salary": [40000, 55000, 35000, 70000, 45000]
}

df = pd.DataFrame(data)
df
```

```python
fig , axes = plt.subplots(1,3 , figsize = (15,5)) 
# telling plt to add 3 subplots in 1 image and setting their size 

axes[0].plot(df["Age"], df["Salary"] , color = 'red', marker ='*' , markersize = 2 , linewidth = 2)
axes[0].grid()
axes[0].set_title("Line plot")
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Salary')

axes[1].hist(df["Salary"] , bins = 5, color = 'green')
axes[1].grid()
axes[1].set_title("Histogram plot")
axes[1].set_xlabel('Salary')
axes[1].set_ylabel('Frequency')


axes[2].boxplot(df["Salary"])
axes[2].grid()
axes[2].set_title("Box plot")
axes[2].set_xlabel('Salary')


plt.show()
```

```python

data1 = {
    "Year": [2019, 2020, 2021, 2022, 2023],
    "Sales": [150000, 180000, 210000, 250000, 300000],
    "Profit": [30000, 35000, 40000, 50000, 65000],
    "Expenses": [120000, 145000, 170000, 200000, 235000]
}

df2 = pd.DataFrame(data1)
df2
```

<br><br>

# Multiplot

```python
plt.plot(df2['Year'], df2['Sales'], label='Sales', marker='o', linewidth=2)
plt.plot(df2['Year'], df2['Profit'], label='Profit', marker='s', linewidth=2)
plt.plot(df2['Year'], df2['Expenses'], label='Expenses', marker='^', linewidth=2)

plt.title("Financial Analysis")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.legend()
plt.show()
```
