```python
import matplotlib.pyplot as plt
```

># Line Plot
* Points connected by lines

* Shows trend, pattern, or change over time

* X-axis = independent variable (time, order)

* Y-axis = dependent variable

```python
# simple syntax
x = [1,2,3,4,5]
y = [2,4,5,7,11]

plt.plot(x, y)
plt.show()
```

```python
# styling line
plt.plot(x, y, color='green', linestyle='--', linewidth=2)
plt.show()
```

```python
# marker on line
plt.plot(x, y, marker='o', color='blue', linewidth=2)
plt.show()
```

```python
# label and title
plt.plot(x, y, marker='o')
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Daily Sales Trend")
plt.grid(True)
plt.show()
```

```python
y2 = [1,3,5,7,9]

plt.plot(x, y, label='Line 1', color='red')
plt.plot(x, y2, label='Line 2', color='blue')

plt.legend()
plt.show()
```

```python
# fill area between line
plt.plot(x, y, color='blue')
plt.fill_between(x, y, color='skyblue', alpha=0.4)
plt.show()
```

## Bar Chart

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.bar(x,y)
plt.show()
```

```python
# can set colour of each bar
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','y','g','b']
plt.bar(x,y, color = c)
plt.show()
```

```python
# do allign horizontaly
import numpy as np

x = np.array([
    'A','B','C','D'
])
y = np.array([23,34,56,78])
plt.barh(x,y, color = 'orange')
plt.show()
```

Colors can be:

'red', 'r'

hex: '#FF0000'

RGB: (0.2, 0.4, 0.6)

```python
# can change the width
x = np.array(['A','B','C','D'])
y = np.array([23,34,56,78])
plt.bar(x,y, color = 'orange' , width = 0.2)
plt.show()
```

```python
# can add title
plt.title("Bar plot")
x = np.array(['A','B','C','D'])
y = np.array([23,34,56,78])
plt.bar(x,y, color = 'orange' , width = 0.2)
plt.show()
```

```python
# can add lables
plt.title("Bar plot")
x = np.array(['A','B','C','D'])
y = np.array([23,34,56,78])
plt.xlabel("Name")
plt.ylabel("Numbers")
plt.bar(x,y, color = 'orange' , width = 0.2)
plt.show()
```

<br><br>

```python
# Bar Edge Styling
plt.bar(x,y,color='green' , edgecolor = 'blue' , linewidth = 2)
plt.show()
```

```python
# transparnecy
plt.bar(x, y, color='blue', alpha=0.4)
plt.show()
```

```python
# Pattern
plt.bar(x, y, color='white', edgecolor='black', hatch='//')
plt.show()
```
