```python
import matplotlib.pyplot as plt
```

### linear graph

```python
x = [5,6,7,8,9,10]
y= [1,2,4,6,8,10]
plt.plot(x,y)
plt.show()       # to remove the detail using show
```

```python
# changing colour
x = [5,6,7,8,9,10]
y= [1,2,4,6,8,10]
c='g'
plt.plot(x,y,c)
plt.show()
```

```python
#to add marker
x = [5,25,15,10,20]
y= [5,34,14,7,20]
plt.plot(x,y,marker = 'o')
plt.show()
```

```python
#to add marker on each axis accordingly
x = [5,25,44,25,12]
y= [5,34,14,7,20]
plt.plot(x,y)
plt.plot(x,marker = '>')
plt.show()
```

## Bar Chart

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.bar(x,y)
plt.show()
```

```python
# can set colour of each bar
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','y','g','b']
plt.bar(x,y, color = c)
plt.show()
```

```python
# do allign horizontaly
import numpy as np

x = np.array([
    'A','B','C','D'
])
y = np.array([23,34,56,78])
plt.barh(x,y, color = 'orange')
plt.show()
```

Colors can be:

'red', 'r'

hex: '#FF0000'

RGB: (0.2, 0.4, 0.6)

```python
# can change the width
x = np.array(['A','B','C','D'])
y = np.array([23,34,56,78])
plt.bar(x,y, color = 'orange' , width = 0.2)
plt.show()
```

```python
# can add title
plt.title("Bar plot")
x = np.array(['A','B','C','D'])
y = np.array([23,34,56,78])
plt.bar(x,y, color = 'orange' , width = 0.2)
plt.show()
```

```python
# can add lables
plt.title("Bar plot")
x = np.array(['A','B','C','D'])
y = np.array([23,34,56,78])
plt.xlabel("Name")
plt.ylabel("Numbers")
plt.bar(x,y, color = 'orange' , width = 0.2)
plt.show()
```

<br><br>

```python
# Bar Edge Styling
plt.bar(x,y,color='green' , edgecolor = 'blue' , linewidth = 2)
plt.show()
```

```python
# transparnecy
plt.bar(x, y, color='blue', alpha=0.4)
plt.show()
```

```python
# Pattern
plt.bar(x, y, color='white', edgecolor='black', hatch='//')
plt.show()
```
