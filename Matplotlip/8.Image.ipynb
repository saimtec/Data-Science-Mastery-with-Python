## Matplotlib can load, display, modify, and annotate images.
**This is super useful for:**

* Computer Vision

* ML / Deep Learning

* Image processing

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread("download.jpeg")
print(img.shape)
```

Meaning:

* 183 rows (height)

* 275 columns (width)

* 3 color channels (RGB)

## Display an Image

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

## Grayscale Image

```python
import numpy as np

img_grey = np.mean(img , axis=2)    #simple sonvert

plt.imshow(img_grey , cmap='grey')
plt.axis('off')
plt.show()
```

## crop image

```python
crop = img[100:150, 60:250]      #[height_start : height_end  , width_s:width_end]
plt.imshow(crop)
plt.axis('off')
plt.show()
```

## plot image over a chart

```python
plt.imshow(img)
plt.scatter([20,30,40],[15,20,25] , color='red' , s=50)
plt.show()
```

## save image after editing

```python
plt.imsave('output.png',img)

# save grey_scale image

plt.imsave("grey_image.png", img_grey , cmap='grey')
```

## colormap

```python
plt.imshow(img_grey, cmap='inferno')
plt.axis('off')
plt.show()
```

## overlay text

```python
plt.imshow(img)
plt.text(50, 50, "Hello!", color='yellow', fontsize=20)
plt.axis('off')
plt.show()
```

**Popular colormaps:**

* gray

* viridis

* inferno

* plasma

* jet

# image data in matplotlib is numpy
* so we can directly change 

```python
bright = img * 0.1
bright = bright.clip(0,1)  # force value to be in between 0 and 1
plt.imshow(bright)
plt.show()
```

# Through PIL (Pillow)

```python
from PIL import Image ,ImageEnhance , ImageFilter
```

```python
fname = r'download.jpeg'
```

```python
# opening image using pil
image = Image.open(fname).convert('L')
```

```python
#mapping image to grey scale
plt.imshow(image , cmap ='gist_heat_r')
plt.show()
```

## Check format

```python
print(image.format)
print(image.size)
print(image.mode)
```

## convert image mode

```python
img = image.convert('RGBA')
print(img.mode)
```

## Adjust brightness, contrast, color, sharpness

```python
enh = ImageEnhance.Brightness(img)
bright = enh.enhance(1.5)  

```

```python
enh = ImageEnhance.Contrast(image)
contrast = enh.enhance(2)
```

```python
enh = ImageEnhance.Color(img)
rich = enh.enhance(1.8)
```

```python
enh = ImageEnhance.Sharpness(img)
sharp = enh.enhance(3)
```

## filter blur image

```python
fname = r'images.jpeg'

images = Image.open(fname).convert('L')
plt.imshow(images)
plt.show()
```

```python
blur = images.filter(ImageFilter.BLUR)
plt.imshow(blur)
plt.show()
```

```python
gauss = images.filter(ImageFilter.GaussianBlur(5))
plt.imshow(gauss)
plt.show()
```

```python
contour = images.filter(ImageFilter.CONTOUR)
contour = images.filter(ImageFilter.DETAIL)
contour = images.filter(ImageFilter.EDGE_ENHANCE)
# contour = images.filter(ImageFilter.SHARPEN)
# contour = images.filter(ImageFilter.EMBOSS)

plt.imshow(contour)
plt.show()
```
