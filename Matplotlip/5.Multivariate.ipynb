```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    # "Name": ["Ali", "Sara", "Zain"],
    "Age": [22, 24, 21,23,24,21],
    "Department": ["IT", "HR", "FINANCE","science","research","Investment"],
    "Experience": [11,10,12,11,9,8],
    "Salary":[100088,100000,122202,100000,90000,80000]
}

df = pd.DataFrame(data)

print(df)
```

# Bublle Plot
* 3 numeric variable

```python
plt.scatter(df['Age'] , df['Salary'] ,s=  df['Experience'] * 20, color='orange' , edgecolor = 'black')
plt.title("Age vs Salary vs Experience")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()

#experieence column as the size
```

```python
plt.scatter(df['Age'] , df['Salary'] ,c = df['Department'].map({'HR':'yellow' , 'IT':'green' , 'FINANCE':'blue' , 'Investment':'orange' , 'research':'gray' , 'science':'black'}), edgecolor = 'black')
plt.title("Age vs Salary vs Experience")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
plt.legend()
#set colours accordingly deparment
```
