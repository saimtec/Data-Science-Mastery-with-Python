```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Profit": [30000, 35000, 40000, 50000, 65000],
    "Age": [22, 24, 21, 27, 23],
    "Salary": [40000, 55000, 35000, 70000, 45000]
}

df = pd.DataFrame(data)
df
```

```python
ax = plt.axes(projection = '3d')

ax.scatter(df["Age"], df["Salary"],df['Profit'])
ax.set_xlabel('Age')
ax.set_ylabel('Salary')
ax.set_yzlabel('Profit')
plt.show()
```
