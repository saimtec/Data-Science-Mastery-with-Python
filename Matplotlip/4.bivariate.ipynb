```python
import matplotlib.pyplot as plt
```

># Scatter Plot

* Points plotted on an X-Y plane

* Each point = (x, y)

* Shows correlation, pattern, clusters, outliers

```python
x = [1,2,3,4,5]
y = [2,4,5,7,11]

plt.scatter(x, y)
plt.show()
```

```python
# Customize Color and Size
plt.scatter(x, y, color='red', s=100)
plt.show()
```

```python
# addomg labels and title
plt.scatter(x, y, color='blue')

plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Study vs Marks")
plt.grid(True)     #Grid helps visually

plt.show()
```

```python
# colour by category
x = [1,2,3,4,5]
y = [2,4,5,7,11]
c = ['red','green','blue','yellow','orange']

plt.scatter(x,y, c=c)
plt.show()
```

```python
# marker style
plt.scatter(x, y, marker='x')
plt.show()     
# o, s, x, *, +, ^, D, v   other styles
```
