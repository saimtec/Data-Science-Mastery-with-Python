># HISTOGRAM
## A histogram shows the distribution of continuous data by grouping values into bins.

**Examples:**

* income distribution

* height distribution

* test scores

* daily returns

```python
import matplotlib.pyplot as plt
import numpy as np

x= [12,13,14,15]
plt.hist(x)
plt.show()
```

```python
# customize bins   (more bins more detail)
plt.hist(x , bins =15)
```

```python
# Colors and Alpha
plt.hist(x, bins=20, color='green', alpha=0.6)
plt.show()
```

```python
# Edge Color
plt.hist(x ,bins = 20 , color='blue' , edgecolor='orange')
plt.show()
```

```python
# with labels and title
plt.hist(x, bins=20, edgecolor='black')

plt.title("Distribution of Test Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# Multiple Histograms (Overlap)
y= [20,30,35,37,40]
z= [20,30,32,34,44]

plt.hist(y ,bins=20, color='orange' , alpha =0.5 ,label='income')
plt.hist(z ,bins=20, color='blue' , alpha =0.5 ,label='spend')
plt.legend()      # show on chart
plt.show()
```

```python
# rotation
plt.hist(x, bins=20, orientation='horizontal')
plt.show()
```

># Pie Chart

* A pie chart divides a circle into slices,
* where each slice represents a categorys share of the total.

* It is based on percentage, not absolute values.

```python
# Basic Pie Chart in Matplotlib
labels = ['A','B','C','D']
sizes = [40,25,20,15]
plt.pie(sizes , labels=labels)
plt.show()
```

```python
# Adding Colors
colour = ['r' , 'g','b','y']
plt.pie(sizes , labels=labels ,colors=colour )
plt.show()
```

```python
# add percent value (autopct)
plt.pie(sizes, labels=labels, autopct ='%1.1f%%')
plt.show()
```

```python
# highlight a category
explode = [0.1,0,0,0]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode)
plt.show()
```

```python
# rotate
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
# adding shadow
plt.pie(sizes , labels =labels , autopct = '%1.1f%%' , shadow = True)
plt.show()
```
